```
import matplotlib.pyplot as plt
import scipy.io.wavfile
import numpy as np
import glob
import scipy.io.wavfile as wf

x = glob.glob('''D:/0_Joowhan's Paper/Synthesized/Total/A3F3P3/A2.503786_F1.078928_P2.372837_H360.032644.wav''')
print(x)

nsc = 1470
nov = nsc/2
nff = nsc 
imgSize = 224

fs, dataInt16 = wf.read('''D:\\0_Joowhan's Paper\\Synthesized\\Total\\A3F3P3\\A2.503786_F1.078928_P2.372837_H360.032644.wav''')
dataFloat = dataInt16 / (2 ** 15)
Pxx, _, _, _ = plt.specgram(dataFloat, NFFT=nff, Fs=fs, noverlap=nov, \
                                   window=np.hamming(nsc), cmap='viridis')
fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)

# plt.colorbar()

plt.clim((-80, -60))

plt.ylim((2000, 6000))

plt.show()
```
