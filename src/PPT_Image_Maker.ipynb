```
import matplotlib.pyplot as plt
import scipy.io.wavfile
import numpy as np
import glob
import scipy.io.wavfile as wf

x = glob.glob('''D:/0_Joowhan's Paper/Real Data/Mando/20171115-133323_IKCK171110A0214-TILTUP.wav''')
print(x)

labelFont = 36
tickFont = 24

nsc = int(1470 / 3)
nov = nsc/2
nff = nsc 
imgSize = 224

fs, dataInt16 = wf.read('''D:/0_Joowhan's Paper/Real Data/Mando/20171115-133323_IKCK171110A0214-TILTUP.wav''')
dataFloat = dataInt16 / (2 ** 15)
Pxx, _, _, _ = plt.specgram(dataFloat[:, 1], NFFT=nff, Fs=fs, noverlap=nov, \
                                   window=np.hamming(nsc), cmap='viridis')
fig = plt.gcf()
fig.set_size_inches(12, 10)
# plt.gca().get_xaxis().set_visible(False)
# plt.gca().get_yaxis().set_visible(False)

cb = plt.colorbar()

cb.ax.set_yticklabels(cb.ax.get_yticklabels(), fontsize=tickFont)

plt.clim((-80, -50))

# plt.ylim((2000, 6000))

plt.xlabel('Time(s)', fontsize=labelFont, labelpad = 24)
plt.xlim([0, 3])
plt.xticks([1 ,2 , 3], fontsize=tickFont)

plt.ylabel('Frequency(Hz)', fontsize=labelFont, labelpad = 24)
plt.xlim([0, 3])
plt.yticks(fontsize=tickFont)


plt.show()

fs, dataInt16 = wf.read('''D:/0_Joowhan's Paper/Real Data/Mando/20180716-071024_IKCK180712A0180-TELEOUT.wav''')
dataFloat = dataInt16 / (2 ** 15)
Pxx, _, _, _ = plt.specgram(dataFloat[int(0.5 * fs): int(3.6 *fs) + 1], NFFT=nff, Fs=fs, noverlap=nov, \
                                   window=np.hamming(nsc), cmap='viridis')
fig = plt.gcf()
fig.set_size_inches(12, 10)
# plt.gca().get_xaxis().set_visible(False)
# plt.gca().get_yaxis().set_visible(False)

cb = plt.colorbar()

cb.ax.set_yticklabels(cb.ax.get_yticklabels(), fontsize=tickFont)

plt.clim((-80, -50))
plt.xlabel('Time(s)', fontsize=labelFont, labelpad = 24)
plt.xlim([0, 3])
plt.xticks([1 ,2 , 3], fontsize=tickFont)

plt.ylabel('Frequency(Hz)', fontsize=labelFont, labelpad = 24)
plt.xlim([0, 3])
plt.yticks(fontsize=tickFont)


# plt.ylim((2000, 6000))

plt.show()
```
